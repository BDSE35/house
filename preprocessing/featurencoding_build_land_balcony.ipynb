```python
df = pd.read_csv('csv/merged_houses_0930w_1001d_1001y_cleaned.csv')
df
```

* '建物型態'做one-hot encoding (型態_...)
* 是否包含陽台'做one-hot encoding(>0就是1)
* 是否包含額外土地'：做one-hot encoding，參照'交易筆棟數'，土地 > 房屋就是1

```python
# 清整建物型態函式

def clean_building_type(row):
    if row['建物型態'] == '透天厝':
        return '透天厝'
    elif row['建物型態'] in ['住宅大樓(11層含以上有電梯)', '辦公商業大樓','華廈(10層含以下有電梯)']:
        return '住商大樓'
    elif row['建物型態'] in ['公寓(5樓含以下無電梯)','套房(1房1廳1衛)']:
        return '公寓'
    else:
        return '其他'

# 新增 '建築型態清整' 欄位
df['建築型態清整'] = df.apply(clean_building_type, axis=1)
```

```python
df['建築型態清整'].value_counts()
```

```python
df['陽台面積'].isna().value_counts()
```

```python
# 是否有陽台函式

def balcony_check(row):
    if row['陽台面積'] > 0:
        return 1
    else:
        return 0

# 新增 '陽台有無' 欄位 有=1,無=0
df['陽台有無'] = df.apply(balcony_check, axis=1)
```

```python
df['陽台有無'].value_counts()
```

```python
df['陽台有無']
```

```python
df['交易筆棟數'].value_counts()
```

```python
# 是否含有額外土地函式

def has_extra_land(row):
    # 解析 '交易筆棟數' 字串
    transaction_info = row['交易筆棟數']
    
    # 刪除車位部分
    transaction_info = transaction_info.split('車位')[0]  # 只保留「土地」和「建物」部分
    
    # 再次解析土地和建物的數量
    parts = transaction_info.split('建物')
    land_count = int(parts[0].replace('土地', '').strip())
    building_count = int(parts[1].strip())  # 直接取建物數量
    
    # 判斷土地筆數和建物筆數的關係
    if land_count > building_count:
        return 1  # 含額外土地
    else:
        return 0  # 不含額外土地

# 新增 '是否含額外土地' 欄位，有=1, 無=0
df['是否含額外土地'] = df.apply(has_extra_land, axis=1)

```

```python
df['是否含額外土地'].value_counts()
```

```python
# 使用 get_dummies 進行 One-Hot Encoding
one_hot_encoded = pd.get_dummies(df['建築型態清整'], prefix='建築型態')

# 將 One-Hot Encoding 的結果與原數據框合併
df = pd.concat([df, one_hot_encoded], axis=1)

df
```

```python
df[['建築型態_住商大樓', '建築型態_公寓', '建築型態_其他', '建築型態_透天厝']] = df[['建築型態_住商大樓', '建築型態_公寓', '建築型態_其他', '建築型態_透天厝']].astype(int)

df
```
