```python
import os
import pandas as pd
from geopy.geocoders import Nominatim
from time import sleep

# 初始化 Nominatim Geocoder
geolocator = Nominatim(user_agent="geoapiExercises")

# 定義一個函式，將地址轉換為經緯度
def geocode_address(address):
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Geocoding error for address {address}: {e}")
        return None, None

# 設定資料夾路徑
csv_dir_path = r"C:/Users/jason/OneDrive/桌面/102-103"

# 讀取四個 CSV 檔案並對每個檔案進行地址轉換
for csv_file in os.listdir(csv_dir_path):
    if csv_file.endswith('.csv'):
        csv_file_path = os.path.join(csv_dir_path, csv_file)
        
        try:
            # 讀取 CSV 檔案
            data = pd.read_csv(csv_file_path)

            # 確保存在 "土地位置建物門牌" 欄位
            if '土地位置建物門牌' in data.columns:
                # 新增兩個欄位來儲存經緯度
                data['Latitude'] = None
                data['Longitude'] = None

                # 逐行處理地址，並轉換為經緯度
                for index, row in data.iterrows():
                    address = row['土地位置建物門牌']
                    latitude, longitude = geocode_address(address)
                    data.at[index, 'Latitude'] = latitude
                    data.at[index, 'Longitude'] = longitude

                    # 避免 API 請求過於頻繁，加入 sleep 延遲
                    sleep(1)

                # 將轉換後的資料儲存回 CSV
                output_file_path = os.path.join(csv_dir_path, f"geocoded_{csv_file}")
                data.to_csv(output_file_path, index=False)
                print(f"已完成 {csv_file} 的經緯度轉換並儲存至 {output_file_path}")
        
        except Exception as e:
            print(f"處理檔案 {csv_file} 時發生錯誤: {e}")
```
